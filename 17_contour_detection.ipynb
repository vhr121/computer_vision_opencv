```python
import cv2
```

```python
img_org = cv2.imread('car.jpg')#read graystyle image
img = cv2.cvtColor(img_org,cv2.COLOR_BGR2GRAY)
```

img = cv2.imread('car.jpg')#read graystyle image


```python
ret, thres_img = cv2.threshold(img,127,255,cv2.THRESH_BINARY)#find the binary threshold
```

```python
cv2.imshow('img',thres_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
contours= cv2.findContours(thres_img,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(img_org, contours[1], -1, (0,255,0), 5)
```
